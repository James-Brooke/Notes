# Plotting side by side with pandas

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
ax1, ax2 = axes.ravel()
```

```python
df = pd.DataFrame(np.random.rand(5,3),
    index=['cust1','cust2','cust3','cust4','cust5'],
    columns=pd.Index(['Metric1','Metric2','Metric3'], name='Metrics'))
```

```python
df.plot(kind='bar', ax=ax1, alpha=0.75, title='Bar plot')
```

```python
plt.setp(ax1.get_xticklabels(), rotation =45, fontsize=10)
plt.setp(ax1.get_yticklabels(), rotation =0, fontsize=10)
```

```python
ax1.set_xlabel('Customer')
ax1.set_ylabel('Value')
ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')
colors = dict(boxes='DarkBlue', whiskers='Gray', medians='Red', caps='Black')
```

```python
df.plot(kind='box', color=colors, sym='r', ax=ax2, title='Box Plot')
```

```python
plt.setp(ax2.get_xticklabels(), rotation =45, fontsize=10)
plt.setp(ax2.get_yticklabels(), rotation =0, fontsize=10)
ax2.set_xlabel('Metric')
ax2.set_ylabel('Value')
ax2.xaxis.set_ticks_position('bottom')
ax2.yaxis.set_ticks_position('left')
```

```python
plt.show()
```

```python
fig.subplots_adjust(wspace=0.4)
```
