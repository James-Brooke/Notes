# Annotating Matplotlib figures

```python
%matplotlib inline
import matplotlib.pyplot as plt 
import matplotlib as mpl
import numpy as np 
```

```python
fig, ax = plt.subplots(figsize=(15,7))

style = dict(size=12, color='black')


#ha = horizontal alignment
ax.text(0.2,0.2, "0.2, 0.2", ha='center', **style)
ax.text(0.4,0.4, "0.4, 0.4", ha='center', **style)
ax.text(0.6,0.6, "0.6, 0.6", ha='center', **style)
ax.text(0.8,0.8, "0.8, 0.8", ha='center', **style);
```

```python
fig, ax = plt.subplots()

x = np.linspace(0, 20, 1000)
ax.plot(x, np.cos(x))
ax.axis('equal')

ax.annotate('local maximum', xy=(6.28, 1), xytext=(10, 4), 
            arrowprops=dict(facecolor='black', shrink=0.05))
            
ax.annotate('local minimum', xy=(5*np.pi, -1), xytext=(2, -6), 
            arrowprops=dict(arrowstyle ="->", connectionstyle="angle3,angleA=0,angleB=-90"));
```
