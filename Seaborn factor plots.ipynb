# Seaborn factor plots

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd 
import numpy as np 
import seaborn as sns
sns.set()
```

```python
#Dataset that can be easilly imported with seaborn. Full list of datasets at: https://github.com/mwaskom/seaborn-data
planets = sns.load_dataset('planets')
planets.head()
```

```python
with sns.axes_style('dark'):
    g = sns.factorplot("year", data=planets, aspect=2,
                      kind="count", color='blue')
    g.set_xticklabels(step=5)
```

```python
with sns.axes_style('dark'):
    g = sns.factorplot("year", data=planets, aspect=3,
                      kind="count", hue='method', order=range(2001, 2015))
    g.set_ylabels('Number of Planets Discovered')
```
