# Adjusting Matplotlib axes limits

```python
%matplotlib inline
import matplotlib.pyplot as plt 
import numpy as np
```

```python
#1000 equally spaced numbers between 0 and 20
x = np.linspace(0,20,1000)
```

```python
plt.plot(x, np.sin(x));
```

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 21)
plt.ylim(-1.5, 1.5)
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 21, -1.5, 1.5]);
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight');
```

```python
#equal aspect ratio so that on your screen, one unit of x is equal to one unit of y
plt.plot(x, np.sin(x))
plt.axis('equal');
```
